```python
# so basically we are fetching the datas separately 
# to fetch in a oragnised order we will create  a entire container of the box
containers=soup.find_all('div',class_='companyCardWrapper')
```

```python
len(containers)
```

```python
name=[]
rating=[]
reviews=[]
ctype=[]
avg_salary=[]
total_jobs=[]
for i in containers :
    name.append(i.find('h2').text.strip())
    rating.append(i.find('div',class_='rating_text').text.strip())
    reviews.append(i.find('span','companyCardWrapper__companyRatingCount').text.strip())
    ctype.append(i.find('span',class_='companyCardWrapper__interLinking').text.strip())
    avg_salary.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[1].text.strip())
    total_jobs.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[3].text.strip())

# the data we have loaded actually for only 1 page
  
```

```python
# to load the entire data from the website 
final=pd.DataFrame()
for j in range(1,501):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
    webpage=requests.get('https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={j}',headers=headers).text
    soup=BeautifulSoup(webpage,'lxml')
    company=soup.find_all('div',class_='companyCardWrapper')

    name=[]
    rating=[]
    reviews=[]
    ctype=[]
    avg_salary=[]
    total_jobs=[]


    for i in company:
        try:
           name.append(i.find('h2').text.strip())
        except:
            name.append(np.nan)

        
        try:
            rating.append(i.find('div',class_='rating_text').text.strip())
        except:
            rating.append(np.nan)

        
        try:
               reviews.append(i.find('span','companyCardWrapper__companyRatingCount').text.strip())
        except:
            reviews.append(np.nan)


        try:
            ctype.append(i.find('span',class_='companyCardWrapper__interLinking').text.strip())
        except:
            ctype.append(np.nan)

        
        try:
            avg_salary.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[1].text.strip())
        except:
            avg_salary.append(np.nan)


        try:
             total_jobs.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[3].text.strip())
        except:
            total_jobs.append(np.nan)

        df=pd.DataFrame({
        'name':name,
        'rating':rating,
        'reviews':reviews,
        'company_type':ctype,
        'salary':avg_salary,
        'jobs':total_jobs
        
        })
    final = pd.concat([final, df], ignore_index=True)
```
